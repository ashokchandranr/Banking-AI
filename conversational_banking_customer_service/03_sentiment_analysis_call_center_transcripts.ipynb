## Step 7 — Model Training & Evaluation

Train the model and compare performance against the baseline.

**Prediction prompt:** Before viewing the results, what performance do you expect and why?

```python
positive_words = ['thank', 'kind', 'love', 'great', 'excellent', 'good']
negative_words = ['frustrated', 'unacceptable', 'delay', 'worst', 'bad', 'terrible']

def score_sentiment(text):
    text = text.lower()
    score = 0
    for w in positive_words: score += text.count(w)
    for w in negative_words: score -= text.count(w)
    return score

df['sentiment_score'] = df['transcript'].apply(score_sentiment)
```

```python
df['sentiment_score'].plot(kind='bar', color=['green' if x >= 0 else 'red' for x in df['sentiment_score']])
plt.title('Sentiment Score of Transcripts')
plt.ylabel('Score')
plt.tight_layout()
plt.show()
plt.close()
```
