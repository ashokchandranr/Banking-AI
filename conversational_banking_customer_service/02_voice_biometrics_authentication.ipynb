## Step 2 — Environment Setup

Import libraries and set deterministic seeds for reproducibility.

```python
# Requires: numpy>=1.24, pandas>=2.0, scikit-learn>=1.3, matplotlib>=3.7, seaborn>=0.13

import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

RANDOM_SEED = 42
RNG = np.random.default_rng(RANDOM_SEED)

print("Environment ready.")
```

## Step 3 — Data Creation & Context

Build Synthetic Dataset

Generate a 'stored' voiceprint for a user and some 'test' voiceprints.

```python

stored_voiceprint = np.random.rand(1, 128) # 128-dimensional embedding

same_user_test = stored_voiceprint + RNG.normal(0, 0.05, (1, 128))
different_user_test = np.random.rand(1, 128)
```

## Step 7 — Model Training & Evaluation

Train the model and compare performance against the baseline.

**Prediction prompt:** Before viewing the results, what performance do you expect and why?

```python
sim_same = cosine_similarity(stored_voiceprint, same_user_test)[0][0]
sim_diff = cosine_similarity(stored_voiceprint, different_user_test)[0][0]

threshold = 0.9
def verify(sim):
    return 'Access Granted' if sim > threshold else 'Access Denied'
```

```python
print(f'Same User Similarity: {sim_same:.4f} -> {verify(sim_same)}')
print(f'Different User Similarity: {sim_diff:.4f} -> {verify(sim_diff)}')
```
