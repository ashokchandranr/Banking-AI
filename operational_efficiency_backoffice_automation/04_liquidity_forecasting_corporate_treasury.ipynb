## Step 2 — Environment Setup

Import libraries and set deterministic seeds for reproducibility.

```python
# Requires: numpy>=1.24, pandas>=2.0, scikit-learn>=1.3, matplotlib>=3.7, seaborn>=0.13

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

RANDOM_SEED = 42
RNG = np.random.default_rng(RANDOM_SEED)
plt.rcParams["figure.figsize"] = (10, 6)

print("Environment ready.")
```

## Step 3 — Data Creation & Context

Build Synthetic Dataset

Synthetic daily cash balance for a year.

```python
days = 365
dates = pd.date_range('2022-01-01', periods=days)
base = 1000000
trend = np.linspace(0, 200000, days)
seasonality = 50000 * np.sin(2 * np.pi * np.arange(days) / 30)
noise = RNG.normal(0, 10000, days)
cash_balance = base + trend + seasonality + noise
df = pd.DataFrame({'date': dates, 'balance': cash_balance})
```

## Step 7 — Model Training & Evaluation

Train the model and compare performance against the baseline.

**Prediction prompt:** Before viewing the results, what performance do you expect and why?

```python
df['day_num'] = np.arange(days)
df['lag_1'] = df['balance'].shift(1)
df.dropna(inplace=True)

X = df[['day_num', 'lag_1']]
y = df['balance']
model = LinearRegression()
model.fit(X, y)
```

```python
last_val = df['balance'].iloc[-1]
last_day = df['day_num'].iloc[-1]
forecast = []
curr_val = last_val
for i in range(1, 11):
    pred = model.predict([[last_day + i, curr_val]])[0]
    forecast.append(pred)
    curr_val = pred

print("10-Day Forecast:", forecast)
plt.plot(range(10), forecast, marker='o')
plt.title('10-Day Liquidity Forecast')
plt.tight_layout()
plt.show()
plt.close()
```

**Alt-text:** Line chart titled "10-Day Liquidity Forecast". The chart shows temporal trends and the alignment between predicted and observed values.
