## Step 2 — Environment Setup

Import libraries and set deterministic seeds for reproducibility.

```python
# Requires: numpy>=1.24, pandas>=2.0, scikit-learn>=1.3, matplotlib>=3.7, seaborn>=0.13

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

RANDOM_SEED = 42
RNG = np.random.default_rng(RANDOM_SEED)
plt.rcParams["figure.figsize"] = (10, 6)

print("Environment ready.")
```

## Step 3 — Data Creation & Context

Build Synthetic Dataset

Generate event logs for 100 customers across 5 stages.

```python
n = 100
stages = ['Application', 'KYC', 'Verification', 'Approval', 'Account Opening']
data = []
for cid in range(1, n + 1):
    current_time = pd.Timestamp('2023-01-01') + pd.Timedelta(days=RNG.integers(0, 30))
    for stage in stages:
        duration = RNG.gamma(shape=2, scale=1.0 if stage != 'Verification' else 5.0) # Verification is slow
        current_time += pd.Timedelta(days=duration)
        data.append({'customer_id': cid, 'stage': stage, 'timestamp': current_time})
df = pd.DataFrame(data)
```

## Step 7 — Model Training & Evaluation

Train the model and compare performance against the baseline.

**Prediction prompt:** Before viewing the results, what performance do you expect and why?

```python
df['prev_timestamp'] = df.groupby('customer_id')['timestamp'].shift(1)
df['duration_days'] = (df['timestamp'] - df['prev_timestamp']).dt.total_seconds() / (24*3600)
bottlenecks = df.groupby('stage')['duration_days'].mean().reindex(stages)
print(bottlenecks)
```

```python
bottlenecks.plot(kind='bar', color='#E76F51', title='Average Duration per Onboarding Stage')
plt.ylabel('Days')
plt.tight_layout()
plt.show()
plt.close()
```
