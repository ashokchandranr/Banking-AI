## Step 2 — Environment Setup

Import libraries and set deterministic seeds for reproducibility.

```python
# Requires: numpy>=1.24, pandas>=2.0, scikit-learn>=1.3, matplotlib>=3.7, seaborn>=0.13

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

RANDOM_SEED = 42
RNG = np.random.default_rng(RANDOM_SEED)
plt.rcParams["figure.figsize"] = (10, 6)

print("Environment ready.")
```

## Step 3 — Data Creation & Context

Build Synthetic Dataset

Synthetic hourly traffic for a branch.

```python
hours = np.arange(9, 18) # 9 AM to 5 PM
days = 30
data = []
for day in range(days):
    for h in hours:
        # Peak at lunch time (12-2 PM)
        base_traffic = 5
        peak = 15 if 12 <= h <= 14 else 0
        noise = RNG.integers(0, 5)
        data.append({'hour': h, 'customers': base_traffic + peak + noise})
df = pd.DataFrame(data)
```

## Step 4 — Exploratory Data Analysis

Exploratory Data Analysis (EDA)

Visualize traffic patterns.

```python
sns.boxplot(x='hour', y='customers', data=df)
plt.title('Hourly Customer Traffic Distribution')
plt.tight_layout()
plt.show()
plt.close()
```

**Alt-text:** Distribution plot titled "Hourly Customer Traffic Distribution". The chart highlights distributional differences between groups that inform the modelling approach.
