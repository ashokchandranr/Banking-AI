## Step 2 — Environment Setup

Import libraries and set deterministic seeds for reproducibility.

```python
# Requires: numpy>=1.24, pandas>=2.0, scikit-learn>=1.3, matplotlib>=3.7, seaborn>=0.13

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

sns.set_theme(style="whitegrid")

RANDOM_SEED = 42
RNG = np.random.default_rng(RANDOM_SEED)
plt.rcParams["figure.figsize"] = (10, 6)

print("Environment ready.")
```

## Step 3 — Data Creation & Context

Build Synthetic Usage Data

We'll create 3 distinct types of customers using different random centers.

```python
# Group 1: Low Balance, High Transactions (Students/Young Professionals)
c1_bal = RNG.normal(2000, 500, 300)
c1_tx = RNG.normal(40, 5, 300)

# Group 2: High Balance, Low Transactions (Wealthy Savers)
c2_bal = RNG.normal(50000, 10000, 300)
c2_tx = RNG.normal(10, 3, 300)

# Group 3: Medium Balance, Medium Transactions (Average Workers)
c3_bal = RNG.normal(15000, 3000, 400)
c3_tx = RNG.normal(25, 5, 400)

df = pd.DataFrame({
    'avg_balance': np.concatenate([c1_bal, c2_bal, c3_bal]),
    'monthly_tx': np.concatenate([c1_tx, c2_tx, c3_tx])
})

plt.figure(figsize=(8, 6))
plt.scatter(df['avg_balance'], df['monthly_tx'], alpha=0.5)
plt.title('Raw Customer Data (Before Clustering)')
plt.xlabel('Average Balance (\$)')
plt.ylabel('Monthly Transactions')
plt.tight_layout()
plt.show()
plt.close()
```

## Step 7 — Model Training & Evaluation

Train the model and compare performance against the baseline.

**Prediction prompt:** Before viewing the results, what performance do you expect and why?

K-Means is sensitive to the scale of numbers (e.g., 50,000 balance vs. 40 transactions). We must scale them first.

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

kmeans = KMeans(n_clusters=3, random_state=RANDOM_SEED, n_init=10)
df['cluster'] = kmeans.fit_predict(X_scaled)

print("Cluster assignment complete.")
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='avg_balance', y='monthly_tx', hue='cluster', data=df, palette='viridis', alpha=0.7)
plt.title('Customer Segments identified by K-Means')
plt.xlabel('Average Balance (\$)')
plt.ylabel('Monthly Transactions')
plt.tight_layout()
plt.show()
plt.close()
```

**Alt-text:** Scatter plot titled "Customer Segments identified by K-Means" with Average Balance (\$) on the x-axis and Monthly Transactions on the y-axis. The chart highlights spatial separation or clustering patterns in the data.

```python
profiles = df.groupby('cluster').mean().round(1)
profiles['name'] = ['Digital/Active', 'Wealthy Saver', 'Middle Class'] # Based on inspection
print(profiles)
```
