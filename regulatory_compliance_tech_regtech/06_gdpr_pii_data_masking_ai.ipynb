## Step 3 — Data Creation & Context

Build Synthetic Dataset

We'll create a list of customer queries containing PII.

```python
queries = [
    "Hi, my name is Alice Smith and I'm having trouble with my account.",
    "Can you send the statement to alice.smith@email.com?",
    "My phone number is 555-0192, please call me back.",
    "I live at 123 Maple Street in London and I lost my card.",
    "This is Bob Jones, my email is bobby.j@provider.net and my number is 555-9988.",
    "The transaction happened while I was in Paris last Tuesday."
]

df = pd.DataFrame({'raw_text': queries})
df
```

## Step 4 — Exploratory Data Analysis

## Step 5 — Feature Engineering

Regex is perfect for structured data like emails and phone numbers.

```python
def mask_pii_regex(text):
    # Mask Emails
    text = re.sub(r'[\w\.-]+@[\w\.-]+\.\w+', '[EMAIL_MASKED]', text)
    # Mask Phone Numbers (simple pattern)
    text = re.sub(r'\d{3}-\d{4}', '[PHONE_MASKED]', text)
    return text

df['regex_masked'] = df['raw_text'].apply(mask_pii_regex)
df[['raw_text', 'regex_masked']]
```

## Step 7 — Model Training & Evaluation

Train the model and compare performance against the baseline.

**Prediction prompt:** Before viewing the results, what performance do you expect and why?

```python
# Simulated NER (Named Entity Recognition)
known_names = ['Alice Smith', 'Bob Jones', 'Alice', 'Bob']
known_locations = ['London', 'Paris', 'Maple Street']

def mask_entities(text):
    for name in known_names:
        text = text.replace(name, '[NAME_MASKED]')
    for loc in known_locations:
        text = text.replace(loc, '[LOCATION_MASKED]')
    return text

df['fully_masked'] = df['regex_masked'].apply(mask_entities)
df[['raw_text', 'fully_masked']]
```

```python
print("Sample Result:")
for i, row in df.iterrows():
    print(f"Original: {row['raw_text']}")
    print(f"Masked:   {row['fully_masked']}")
    print("-" * 30)
```
